```python
plt.plot(np.sin(samples));
plt.plot(np.cos(samples));
plt.savefig('sincos.png')
```

```python
raw_image = cv.imread('sincos.png',1) #1 means color are preserved
```

```python
raw_image.shape
```

```python
plt.imshow(raw_image);
```

```python
plt.imshow(raw_image[:,:,0]);#showing red channel
```

```python
plt.imshow(raw_image[:,:,1]);#showing green channel
```

```python
plt.imshow(raw_image[:,:,2]);#showing blue channel and only one pulse is visible as blue color has (0,0,255) and white color has(255,255,255) so in blue channel backgrnd is same
```

```python
vertically_stacked_image = np.vstack((raw_image,raw_image))
```

```python
plt.imshow(vertically_stacked_image);
cv.imwrite('vertically_stacked_imgae.png',vertically_stacked_image)
```

```python
horizontally_stacked_image = np.hstack((vertically_stacked_image,vertically_stacked_image))
```

```python
cv.imwrite('horizontally_stacked_image.png',horizontally_stacked_image)
```
